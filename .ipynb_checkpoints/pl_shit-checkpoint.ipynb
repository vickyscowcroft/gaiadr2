```python
def grab_extinction(row):
    star = row.ID
    ra = row.ra_2
    dec = row.dec_2
    coord_string = str(ra) + 'd ' +  str(dec) + 'd'
    C = coord.SkyCoord(coord_string, frame='fk5')
    table = IrsaDust.get_extinction_table(C)
    irac_1_sandf = table[19][3]
    irac_2_sandf = table[20][3]
    dr2_df.ix[dr2_df.ID==star, 'A_3p6'] = irac_1_sandf
    dr2_df.ix[dr2_df.ID==star, 'A_4p5'] = irac_2_sandf
    print (star, irac_1_sandf, irac_2_sandf)
    return(0)
 
```

```python
dr2_df.apply(lambda line: grab_extinction(line), axis=1);
```

# Use Astrometric Based Luminosities to get distances from Parallaxes

```python
def abl(row):
    star = row.ID
    plx = row.parallax
    mag1 = row.m3p6
    a1 = row.A_3p6
    mag2 = row.m4p5
    a2 = row.A_4p5
    
    dr2_df.ix[dr2_df.ID==star, 'ABL_3p6'] = (plx*10**-3)*10**(0.2*(mag1 - a1) - 2.0)
    dr2_df.ix[dr2_df.ID==star, 'ABL_4p5'] = (plx*10**-3)*10**(0.2*(mag2 - a2) - 2.0)
    
```

```python
dr2_df.apply(lambda line: abl(line), axis=1);
```

```python
dr2_df['abs_mag_3p6'] = dr2_df.m3p6 - dr2_df.A_3p6 - 5*np.log10(1./(dr2_df.parallax*10**-3)) + 5.
dr2_df['abs_mag_4p5'] = dr2_df.m4p5 - dr2_df.A_4p5 - 5*np.log10(1./(dr2_df.parallax*10**-3)) + 5.
```
